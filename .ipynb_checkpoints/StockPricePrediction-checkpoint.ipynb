```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use('fivethirtyeight')
```

```python
# Import stock price of Apple
df = pd.read_csv('./data/AAPL.csv')
df = df.set_index(pd.DatetimeIndex(df['Date'].values))

df['Rolling 7 days'] = df['Close'].rolling(7).mean()
df['Rolling 30 days'] = df['Close'].rolling(30).mean()
```

```python
cols = ['Close', 'Rolling 7 days', 'Rolling 30 days']

plt.figure(figsize=(16, 5))
for col in cols:
    plt.plot(df[col], label=col)
plt.xlabel('Date')
plt.ylabel('USD [$]')
plt.title('Stock Price Apple')
plt.legend()
```
